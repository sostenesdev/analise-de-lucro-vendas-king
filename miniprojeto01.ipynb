```python
import pandas as pd
import folium
from folium.plugins import HeatMap
import matplotlib.pyplot as plt
import seaborn as sns
url ="https://raw.githubusercontent.com/sostenesdev/analise-de-lucro-vendas-king/main/kc_house_data.csv"
# Ts = pd.read_csv('kc_house_data.csv', squeeze = True)
Ts = pd.read_csv(url, squeeze = True)
Ts.isnull().sum()
df = pd.DataFrame(Ts)

#agrupando os imóveis por id
grouped = df.groupby('id')

outputlist = []
for imovel_id, group in grouped :
    #obtendo as linhas de cada imóvel ordenados por data da mais nova para mais antiga (importante para calculo do lucro)
    imovel = df[df["id"] == imovel_id].sort_values(by="date", ascending=False).copy()
    lucro = 0
    #calculando o lucro de cada imovel
    lucro = imovel.iloc[0]['price']*100/( imovel.iloc[-1]['price'] -100)
    #adicionando resultado do lucro em uma lista
    outputlist.append([imovel_id, lucro,imovel.iloc[0]['lat'],imovel.iloc[0]['long']])
lucro_df = pd.DataFrame(outputlist, columns = ['id', 'lucro','lat', 'long'])
output = []
lucro_df.sort_values(by="lucro", ascending=False)

m2 = folium.Map(location=[47.58766, -121.83072], zoom_start=9,tiles='Stamen Terrain')

lat = lucro_df.lat.tolist()
lng = lucro_df.long.tolist()
lucro = lucro_df.lucro.tolist()


#heatmap do lucro por imóvel
HeatMap(list(zip(lat, lng, lucro))).add_to(m2)
m2



```
